```python
data_f = pd.read_csv('/content/InteractiveSheet_2025-06-13_07_39_43 - Лист1.csv')
```

```python
data_f.head(40)
```

```python
data_f['correct_answer'].str.startswith("В тексте нет информации").sum()
```

```python
from bert_score import score
import numpy as np
# Фильтрация пустых значений
df_eval = data_f.dropna(subset=['correct_answer', 'generated_answer'])

# Списки текстов
references = df_eval['correct_answer'].astype(str).tolist()
candidates = df_eval['generated_answer'].astype(str).tolist()
P, R, F1 = score(candidates, references, lang="ru")  # или "en", если тексты на английском

print(f"BERTScore Precision:  {P.mean():.4f}")
print(f"BERTScore Recall:     {R.mean():.4f}")
print(f"BERTScore F1-score:   {F1.mean():.4f}")

```

```python
import matplotlib.pyplot as plt

# Значения метрик
metric_names = ['Precision', 'Recall', 'F1']
metric_values = [P.mean(), R.mean(), F1.mean()]

plt.figure(figsize=(6, 4))
bars = plt.bar(metric_names, metric_values, color='#4A3AFF')
plt.title('Средние значения BERTScore')
plt.ylim(0, 1)
plt.ylabel('Значение')
plt.gca().spines[['top', 'right']].set_visible(False)

# Добавим подписи над столбцами
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.01, f'{height:.4f}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
